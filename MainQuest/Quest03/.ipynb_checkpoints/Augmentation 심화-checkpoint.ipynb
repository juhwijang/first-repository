```python
%%time
import os
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.models as models
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from torchvision import transforms
from torchvision.datasets import ImageFolder

# =========================================================================
# [기본 환경 설정] 경로 및 하이퍼파라미터 (★여기서 재료를 만듭니다★)
# =========================================================================
dataset_dir = '/home/jovyan/work/data_augmentation/data/Images'
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
EPOCH = 3

# 만약 대문자 Images 폴더가 없으면 소문자 images 폴더 확인
if not os.path.exists(dataset_dir):
    dataset_dir = '/home/jovyan/work/data_augmentation/data/images'

print(f"현재 데이터셋 경로: {dataset_dir}")

try:
    # [핵심] 원본 데이터셋(ds_train, ds_test) 역할을 할 베이스를 강제로 생성합니다.
    # 이 부분이 누락되어 그동안 계속 NameError가 났던 것입니다.
    full_dataset = ImageFolder(root=dataset_dir)
    
    # 훈련용과 테스트용 데이터셋 분리 (8:2 비율 예시)
    train_size = int(0.8 * len(full_dataset))
    test_size = len(full_dataset) - train_size
    ds_train, ds_test = torch.utils.data.random_split(full_dataset, [train_size, test_size])
    
    num_classes = len(full_dataset.classes)
    print(f"✅ 원본 데이터 뼈대 생성 완료! 클래스 수: {num_classes}")
except Exception as e:
    print(f"❌ 데이터 로드 실패: {e}\n압축 해제가 올바르게 안 되었거나 경로가 다를 수 있습니다.")

# =========================================================================
# [전처리 함수] 데이터셋에 변환 및 로더 적용
# =========================================================================
def apply_normalize_on_dataset(dataset, with_aug=False, is_test=False):
    base_transforms = [
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ]
    if with_aug and not is_test:
        aug_transforms = [
            transforms.RandomResizedCrop(224),
            transforms.RandomHorizontalFlip(),
            transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2)
        ]
        final_transform = transforms.Compose(aug_transforms + base_transforms[1:])
    else:
        final_transform = transforms.Compose(base_transforms)
        
    # Subset 객체와 일반 Dataset 객체 모두에 대응하도록 변환 적용
    if hasattr(dataset, 'dataset'):
        dataset.dataset.transform = final_transform
    else:
        dataset.transform = final_transform
        
    return DataLoader(dataset, batch_size=32, shuffle=not is_test)

# =========================================================================
# 1단계: 드디어 데이터로더 세팅 (여기서 변수가 확실하게 등록됩니다)
# =========================================================================
ds_train_no_aug = apply_normalize_on_dataset(ds_train, with_aug=False)
ds_train_aug = apply_normalize_on_dataset(ds_train, with_aug=True)
ds_test_loader = apply_normalize_on_dataset(ds_test, is_test=True)
print("✅ 1단계 완료: 데이터셋 변수 할당 성공!")

# =========================================================================
# 2단계: 이미지 2개 꺼내서 시각화 확인하기
# =========================================================================
print("\n--- 2단계: 샘플 이미지 시각화 ---")
data_iter = iter(ds_train_no_aug)
images, labels = next(data_iter)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))
img1 = images[0].permute(1, 2, 0).cpu().numpy() * 0.25 + 0.5
axes[0].imshow(img1.clip(0, 1))
axes[0].set_title(f"Sample 1 (Label: {labels[0].item()})")
axes[0].axis('off')

img2 = images[1].permute(1, 2, 0).cpu().numpy() * 0.25 + 0.5
axes[1].imshow(img2.clip(0, 1))
axes[1].set_title(f"Sample 2 (Label: {labels[1].item()})")
axes[1].axis('off')
plt.show()

# =========================================================================
# 3단계: 쌍둥이 모델 생성 및 구조 변경 (전이 학습)
# =========================================================================
resnet50 = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
resnet50.fc = nn.Linear(resnet50.fc.in_features, num_classes)

aug_resnet50 = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
aug_resnet50.fc = nn.Linear(aug_resnet50.fc.in_features, num_classes)
print("✅ 3단계 완료: 쌍둥이 모델 세팅 완료!")

# =========================================================================
# 4단계: 공통 학습(train) 함수 정의
# =========================================================================
criterion = nn.CrossEntropyLoss()

def train(model, train_loader, test_loader, epochs):
    model.to(device)
    current_optimizer = optim.SGD(model.parameters(), lr=0.001)
    history = {'val_accuracy': []}

    for epoch in range(epochs):
        model.train()
        correct, total = 0, 0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            current_optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            current_optimizer.step()
            
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

        train_acc = 100. * correct / total
        print(f"Epoch [{epoch+1}/{epochs}], Train Accuracy: {train_acc:.2f}%")

        model.eval()
        correct, total = 0, 0
        with torch.no_grad():
            for images, labels in test_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                _, predicted = outputs.max(1)
                total += labels.size(0)
                correct += predicted.eq(labels).sum().item()

        val_acc = 100. * correct / total
        history['val_accuracy'].append(val_acc)
        print(f"Validation Accuracy: {val_acc:.2f}%")

    return history

# =========================================================================
# 5단계: 모델 학습 시작 및 그래프 결과 시각화
# =========================================================================
print("\n--- 4단계 [A]: 일반 모델(No Augmentation) 훈련 시작 ---")
history_resnet50_no_aug = train(resnet50, ds_train_no_aug, ds_test_loader, EPOCH)

print("\n--- 4단계 [B]: 증강 모델(With Augmentation) 훈련 시작 ---")
history_resnet50_aug = train(aug_resnet50, ds_train_aug, ds_test_loader, EPOCH)

print("\n--- 5단계: 최종 성능 비교 그래프 출력 ---")
plt.figure(figsize=(10, 6))
plt.plot(history_resnet50_no_aug['val_accuracy'], 'r', marker='o', label='No Augmentation')
plt.plot(history_resnet50_aug['val_accuracy'], 'b', marker='s', label='With Augmentation')
plt.title('Model Validation Accuracy Comparison')
plt.ylabel('Accuracy (%)')
plt.xlabel('Epoch')
plt.legend(loc='upper left')
plt.grid(True)
plt.ylim(0.50, 100.0)
plt.show()
```

```python
def get_clip_box(image_a, image_b):
    # image.shape = (channel, height, width)
    image_size_x = image_a.shape[2]  # Width
    image_size_y = image_a.shape[1]  # Height

    # Get center of box
    x = torch.randint(0, image_size_x, (1,)).item()
    y = torch.randint(0, image_size_y, (1,)).item()

    width = max(1, int(image_size_x * torch.sqrt(1 - torch.rand(1)).item()))
    height = max(1, int(image_size_y * torch.sqrt(1 - torch.rand(1)).item()))

    # Clip box in image and get minmax bbox
    x_min = max(0, x - width // 2)
    y_min = max(0, y - height // 2)
    x_max = min(image_size_x, x + width // 2 + 1)
    y_max = min(image_size_y, y + height // 2 + 1)

    return x_min, y_min, x_max, y_max

x_min, y_min, x_max, y_max = get_clip_box(image_a, image_b)

print('x :', x_min, x_max)
print('y :', y_min, y_max)
```
