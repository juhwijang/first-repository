```python
import os

data_dir = os.path.join(os.getenv("HOME"), "work/data")
file_name = os.path.join(data_dir, "synopsis.txt")

with open(file_name, 'r') as file:
    for i in range(20):
        print(file.readline(), end='')
```

```python
# 약 15분정도 걸립니다.
from konlpy.tag import Okt

okt = Okt()
tokenized = []

with open(file_name, 'r') as file:
    while True:
        line = file.readline()
        if not line: break
        words = okt.pos(line, stem=True, norm=True)
        res = []
        for w in words:
            if w[1] in ["Noun"]:      # "Adjective", "Verb" 등을 포함할 수도 있습니다.
                res.append(w[0])    # 명사일 때만 tokenized 에 저장하게 됩니다.
        tokenized.append(res)

print("슝~")
```

```python
from gensim.models import Word2Vec

# tokenized에 담긴 데이터를 가지고 나만의 Word2Vec을 생성합니다. (Gensim 4.0 기준)
model = Word2Vec(tokenized, vector_size=100, window=5, min_count=3, sg=0)
model.wv.most_similar(positive=['가수'])

# Gensim 3.X 에서는 아래와 같이 생성합니다.
# model = Word2Vec(tokenized, size=100, window=5, min_count=3, sg=0)
# model.most_similar(positive=['영화'])
```

```python
# [TF-IDF 문제점]
# 1. 문서 길이 불균형의 한계: 특정 장르(문서)의 텍스트 양이 너무 많으면, 해당 장르에서 자주 나오는 흔한 단어가 불리하게 평가되거나 과도하게 노출될 수 있음.
# 2. 장르 간 중복 단어 문제: 여러 장르에 걸쳐 공통적으로 자주 등장하는 범용적인 명사가 각 장르의 핵심 '대표 단어'로 잘못 선정되어 WEAT 스코어의 변별력을 떨어뜨림.
# 3. 문맥 무시: 단어의 등장 빈도만 계산하므로, '가수'라는 키워드와 실제 의미론적으로 밀접한 단어인지 판별하지 못함.
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# 1. tokenized 내부의 단어 리스트들을 공백으로 이어진 하나의 문장들로 변환합니다.
# ['가수', '음악', '무대'] -> "가수 음악 무대"
tokenized_sentences = [" ".join(words) for words in tokenized]

# 2. TF-IDF 학습
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(tokenized_sentences)
features = np.array(vectorizer.get_feature_names_out())

print("TF-IDF 행렬 생성 완료!")
```

```python
# [개선된 방법: TF-IDF 상위 단어 중 지나치게 흔한 단어 제외하기]
# 전체 문서에서 점수가 지나치게 높은 상위 범용 단어는 중복 제거
sums = tfidf_matrix.sum(axis=0).A1
top_indices = sums.argsort()[::-1]

# 너무 자주 나오는 상위 20개 단어는 제외 리스트
stop_words = [features[i] for i in top_indices[:20]] 

# ------------------------------------------------------------------
# [나만의 '가수' 중심 단어 셋 구성]
# 원래 model.wv.most_similar('가수') 결과나 TF-IDF 상위 단어를 참고하여 셋을 구성합니다.
# ------------------------------------------------------------------

# Target A: 가수/음악 연관 핵심 단어 (중복 단어 제외)
target_A = [w for w in ['가수', '노래', '음악', '무대', '앨범', '공연', '아이돌', '뮤지컬', '콘서트', '팬'] if w not in stop_words]

# Target B: 가수와 거리가 먼 사극/전쟁 관련 단어
target_B = [w for w in ['조선', '왕', '전쟁', '장군', '백성', '궁궐', '전투', '군사', '역사', '황제'] if w not in stop_words]

# Attribute X: 긍정적 속성 단어
attribute_X = ['사랑', '최고', '감동', '행복', '환호', '인기', '성공', '기쁨', '열정', '아름답다']

# Attribute Y: 부정적 속성 단어
attribute_Y = ['슬픔', '실패', '이별', '갈등', '외면', '상처', '눈물', '위기', '절망', '고통']

print("A :", target_A)
print("B :", target_B)
```

```python
from numpy import dot
from numpy.linalg import norm

# 코사인 유사도를 구하는 함수 정의
def cos_sim(A, B):
    return dot(A, B)/(norm(A)*norm(B))

# WEAT 공식 중 분자의 단일 단어 편향성 s(w, X, Y) 계산 함수
def s_word(w, X, Y):
    mean_X = np.mean([cos_sim(model.wv[w], model.wv[x]) for x in X if x in model.wv])
    mean_Y = np.mean([cos_sim(model.wv[w], model.wv[y]) for y in Y if y in model.wv])
    return mean_X - mean_Y

# 최종 WEAT score 계산 함수 (효과 크기)
def weat_score(A, B, X, Y):
    s_A = [s_word(a, X, Y) for a in A if a in model.wv]
    s_B = [s_word(b, X, Y) for b in B if b in model.wv]
    
    mean_A = np.mean(s_A)
    mean_B = np.mean(s_B)
    
    # 분모에 들어갈 표준편차 계산
    std_dev = np.std(s_A + s_B)
    
    return (mean_A - mean_B) / std_dev

# 계산 실행
score = weat_score(target_A, target_B, attribute_X, attribute_Y)
print(f" WEAT Score: {score:.4f}")
```

```python
from numpy import dot
from numpy.linalg import norm
import numpy as np

# 코사인 유사도
def cos_sim(A, B):
    return dot(A, B)/(norm(A)*norm(B))

# 단어 하나에 대한 편향성 점수 s(w, X, Y)
def s_word(w, X, Y):
    X_w = [cos_sim(model.wv[w], model.wv[x]) for x in X if x in model.wv]
    Y_w = [cos_sim(model.wv[w], model.wv[y]) for y in Y if y in model.wv]
    if not X_w or not Y_w:
        return 0
    return np.mean(X_w) - np.mean(Y_w)

# 최종 WEAT score (X, Y 고정 / A, B 변동)
def weat_score(X, Y, A, B):
    s_A = [s_word(a, X, Y) for a in A if a in model.wv]
    s_B = [s_word(b, X, Y) for b in B if b in model.wv]
    if not s_A or not s_B:
        return 0
    
    std_dev = np.std(s_A + s_B)
    if std_dev == 0:
        return 0
    return (np.mean(s_A) - np.mean(s_B)) / std_dev
```
